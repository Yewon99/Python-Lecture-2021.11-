- 세자리수 두개를 곱해서 만든 수가 대칭수(palindrome)일 때
- 가장 큰 대칭수와 그 때의 두 수를 구하시오.

```python
import math
two = math.sqrt(4)
two
```

```python
def is_palindrome(n):
    s = str(n)
    return s == s[::-1]
```

```python
# is_palindrome(12321)
```

```python
for i in range(100, 1000):
    for k in range(i, 1000):   # 곱셈의 교환법칙으로 i에서 시작
        if is_palindrome(i*k):
            print(i*k, end=' ')
```

```python
# 상용의 기법 (sorting을 하기 위한 방법 중 하나)
max_num = 0   # max_num에 위에 나온 대칭수를 대입
for i in range(100, 1000):
    for k in range(i, 1000):
        mul = i * k
        if is_palindrome(mul):
            if mul > max_num:
                max_num = mul
                num1, num2 = i, k
print(max_num, num1, num2)
```

- 모든 경우의 수를 다 따지기 때문에 반복문을 사용
